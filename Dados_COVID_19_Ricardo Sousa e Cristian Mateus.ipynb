```python
df.groupby('ID_MUNICIP').size().sort_values().tail(7).plot(kind='bar',color= '#AB2D12',edgecolor='black')
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (21, 21) #Tamanho
plt.title('Cidades com maiores casos',fontsize=22 ) #TÃ­tulo do grÃ¡fico
plt.xticks(rotation=45) # Rotacionar os nomes das cidades
plt.xlabel('')
plt.show()
```

```python
df.groupby('CS_SEXO').size().sort_values().tail(3).plot(kind='barh', color= '#1F7056',edgecolor='black')
plt.rcParams['figure.figsize'] = (10, 10)
plt.title('AnÃ¡lise de casos por sexo',fontsize=22)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.ylabel('')
plt.show()
```

```python
df.groupby('NU_IDADE_N').size().plot(kind='line').set_title('AnÃ¡lise de casos por idade', fontsize = 21)
plt.xlabel('')
plt.show()
```

```python
df.groupby('CS_ETINIA').size().sort_values().tail(7).plot(kind='bar',color= '#7E3005',edgecolor='black')
plt.rcParams['figure.figsize'] = (21, 21) #Tamanho
plt.title('Casos nas principais etnias indÃ­genas',fontsize=22 ) #TÃ­tulo do grÃ¡fico
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right') # Rotacionar os nomes das etnias
plt.xlabel('')
plt.show()
```
